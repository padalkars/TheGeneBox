```python
def process_snp(ind):
    #The changes made in the result will be reflected globally
    global result_df
    for grp in range(1, 32):
        #genotype_list = ['AA', 'AB', 'BB']

        #Search, which column from S_col_list in the group contains a 1
        for i in range(len(genotype_list)):
            #Columns are not unique in the data
            #Hence iterating upon index of genotype_list
            
            res = ''
            if(data.iloc[ind, (3*grp) + i]==1):
                
                ele = genotype_list[i]
                first, second = ele[0], ele[1]
                res += data.loc[ind, first]
                res += data.loc[ind, second]
                break
            
            
        #Update the entry in the resultant data frame
        result_df.loc[ind, 'SNP'] = data.loc[ind, 'SNP']
        result_df.loc[ind, 'S_' + str(grp)] = res
```

```python
#Apply the function 'process_snp' to each 'snp'
list(map(process_snp, data.index))
```
